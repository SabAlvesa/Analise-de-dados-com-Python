```python

lista_notas = [ 8, 9, 10, 7, 4]

print("Valores da notas: ")
for lista in lista_notas:
  print(f"Nota: {lista}")

def calcular_media(notas):

    soma_notas = sum(notas)


    media = soma_notas / 5

    return media

print(f"A m√©dia das notas √©: {calcular_media(lista_notas)}")
```

#### **Exerc√≠cio 5**

Com base no dicion√°rio a seguir, encontre e retorne uma lista de todas as chaves que tem o valor `True`

```python
# Exerc√≠cio 5
valor = True
dados_loja = {
    'nome': 'Loja de Livros Fant√°sticos',
    'faturamento': 150000.00,
    'estoque_baixo': True,
    'cidades_atendidas': ['S√£o Paulo', 'Curitiba', 'Cuiab√°', 'Manaus', 'Recife'],
    'meta_atingida': False,
    'ano_fundacao': 2010,
    'loja_online': True
}
def buscar_valor(input_dict, input_valor):

  chaves_verdadeiras = []

  for k, v in input_dict.items():

    if v == input_valor:
      chaves_verdadeiras.append(k)
  return chaves_verdadeiras

# Exibindo o resultado do exerc√≠cio 5
print(buscar_valor(dados_loja, valor))
```

```python
dicionario_teste = { 'valor1': True, 'valor2': False}
assert buscar_valor(dicionario_teste, True) == ['valor1']
print('Teste passou!')
```
